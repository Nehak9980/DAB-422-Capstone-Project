## Importing required libraries

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore")
```

### EDA and Univariate Analysis

```python
age_data = pd.read_excel('age_data.xlsx')

sns.set_style("whitegrid")

# Summary of categorical variables
print("Unique Age Groups:", age_data['AGE GROUP'].unique())
print("Unique Provinces:", age_data['Province'].unique())

# Count of records for each age group and province
age_group_counts = age_data['AGE GROUP'].value_counts()
province_counts = age_data['Province'].value_counts()

# Trends over the years for different age groups
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='Total', hue='AGE GROUP', data=age_data, palette="pastel", marker='o')
plt.title('Trends Over the Years for Different Age Groups')
plt.ylabel('Total')
plt.xlabel('Year')
plt.legend(title='Age Group', loc='upper left')
plt.show()

```
